```python
fronius_url = "http://192.168.178.70/"
```

```python
import requests
```

```python
# get battery status

url = fronius_url + "solar_api/v1/GetStorageRealtimeData.cgi"

res = requests.get(url)
```

```python
res.json()['Body']['Data']['0']['Controller']
```

* Voltage_DC

# Inverter requests

## Inverter system-scoped requests

```python
# inverter system scope:

inv_system_scope_url = fronius_url + "solar_api/v1/GetInverterRealtimeData.cgi"

# requests parameters - defaults
params = {
    "Scope" : "System"
}

res = requests.get(inv_system_scope_url)
print(res.status_code)
```

```python
res.json()['Body']
```

## Inverter device-scoped requests

```python
url = fronius_url + "solar_api/v1/GetInverterRealtimeData.cgi"

# requests parameters - defaults
params = {
    "Scope" : "Device",
    "DeviceId" : "1"
}
```

```python
# common inverter data
p  = dict({"DataCollection" : "CommonInverterData"}, **params)

res = requests.get(url, params=p)
print(res.status_code)
```

```python
res.json()['Body']
```

```python
# 3PInverterData
p  = dict({"DataCollection" : "3PInverterData"}, **params)

res = requests.get(url, params=p)
print(res.status_code)
```

```python
res.json()['Body']
```
